#### Use the iris database to answer the following quesitons:

2. Is there a correlation between petal length and petal width?

```python
iris_ds = sns.load_dataset('iris')
```

```python
#get to know the data:
iris_ds
```

### #1 What does the distribution of petal lengths look like?

```python
sns.histplot(data = iris_ds, x='petal_length')
```

### #2 Is there a correlation between petal length and petal width?
- Answer: yes, generally, the longer the petal, the wider the petal

```python
sns.histplot(data = iris_ds, x='petal_length', y='petal_width')
```

### #3 Would it be reasonable to predict species based on sepal width and sepal length?
- Answer: it depends/ yes and no. 
    - Virginicia have longer sepal lengths and are wider
    - Setosa have shorter sepal lengths
    - Versicolor are the shortest and narrowest
    - BUT there are some outliers

```python
sns.histplot(data = iris_ds, x='sepal_width', y='sepal_length', hue ='species')
```

### #4 Which features would be best used to predict species?
- Answer: petal_length

```python
#look at data sorted different ways
iris_ds.sort_values('petal_length')
```

```python
#petal_length could best predict species
sns.histplot(data = iris_ds, x='petal_length', y='petal_width', hue = 'species')
```

```python
#this gives average petal length
iris_ds.groupby('species').petal_length.agg('mean')
```
